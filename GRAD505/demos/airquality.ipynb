```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
airquality = pd.read_csv("http://www.brycejdietrich.com/files/airquality.csv")
import calendar
```

```python
print(airquality)
```

```python
print('Mean Ozone: ' + str(airquality['Ozone'].mean()))
print('Min Wind Speed in July: ' + str(airquality['Wind'][airquality['Month']==7].min()))
print('Ozone on Highest Temp: ' + str(airquality['Ozone'][airquality['Temp'] == airquality['Temp'].max()]))
```

```python
plt.plot(airquality.index+1, airquality['Temp'])
```

```python
airquality['MonthName'] = airquality['Month'].apply(lambda x: calendar.month_name[x])
print(airquality)
```

```python
def FtoC(fahrenheit):
    return (fahrenheit - 32) * 5/9
airquality['TempC'] = FtoC(airquality['Temp'])
airquality.head()
```

```python
print(FtoC(40))
print(FtoC(0))
print(FtoC(-40))
```

```python
airquality = airquality.drop(columns=['MonthName', 'TempC'])
airquality.head()
```
